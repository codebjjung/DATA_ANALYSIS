```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1+ np.exp(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)

fig = plt.figure(figsize=(8,7))
fig.set_facecolor('white')

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.xlim(-10, 10)
plt.title("Sigmoid", fontsize = 30)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', rotation=0, fontsize=20)

plt.yticks([0.0, 0.5, 1.0])
plt.axvline(0.0, color='k')
ax = plt.gca()
ax.yaxis.grid(True)

plt.show()
```

```python
def tanh(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = tanh(x)

fig = plt.figure(figsize=(8,7))
fig.set_facecolor('white')

plt.plot(x, y)
plt.ylim(-1.1, 1.1)
plt.xlim(-10, 10)
plt.title("tanh", fontsize = 30)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', rotation=0, fontsize=20)

plt.yticks([-1, 0, 1])
plt.axvline(0.0, color='k')
ax = plt.gca()
ax.yaxis.grid(True)

plt.show()
```

```python
def ReLU(x):
  return np.maximum(0,x)

x = np.arange(-10.0, 10.0, 0.1)
y = ReLU(x)

fig = plt.figure(figsize=(8,7))
fig.set_facecolor('white')

plt.plot(x, y)
plt.ylim(-0.1, 10)
plt.xlim(-10, 10)
plt.title("ReLU", fontsize = 30)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', rotation=0, fontsize=20)

plt.yticks([0, 5, 10])
plt.axvline(0.0, color='k')
ax = plt.gca()
ax.yaxis.grid(True)

plt.show()
```

```python
def Leaky_ReLU(x):
  return np.maximum(0.01*x,x)

x = np.arange(-10.0, 10.0, 0.1)
y = Leaky_ReLU(x)

fig = plt.figure(figsize=(8,7))
fig.set_facecolor('white')

plt.plot(x, y)
plt.ylim(-1, 10)
plt.xlim(-10, 10)
plt.title("Leaky_ReLU", fontsize = 30)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', rotation=0, fontsize=20)

plt.yticks([0, 5, 10])
plt.axvline(0.0, color='k')
ax = plt.gca()
ax.yaxis.grid(True)

plt.show()
```
